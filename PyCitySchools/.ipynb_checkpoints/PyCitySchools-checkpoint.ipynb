```python
# Dependencies and Setup
import pandas as pd

# Files
school_data_file = "Resources/schools_complete.csv"
student_data_file = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_file)
student_data = pd.read_csv(student_data_file)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete
```

```python
# DISTRICT SUMMARY

# Calculate the total number of schools
school_data_df = school_data_complete.copy()
total_schools = school_data_df["school_name"].nunique()


# Calculate the total number of students
total_students = school_data_df["Student ID"].nunique()


# Calculate the total budget
total_budget = sum(school_data_df["budget"].unique())


# Calculate the average math score
avg_math_score = school_data_df["math_score"].mean()


# Calculate the average reading score
avg_reading_score = school_data_df["reading_score"].mean()


# Calculate the percentage of students with a passing math score (70 or greater)
num_passing_math = school_data_df.loc[school_data_df['math_score'] >= 70]
percent_passing_math = len(num_passing_math)/ total_students * 100

# Calculate the percentage of students with a passing reading score (70 or greater)
num_passing_reading = school_data_df.loc[school_data_df['reading_score'] >= 70]
percent_passing_reading = len(num_passing_reading)/ total_students * 100


# Calculate the percentage of students who passed math AND reading (% Overall Passing)
overall_passing = school_data_df.loc[(school_data_df['reading_score'] >= 70) & (school_data_df['math_score'] >= 70)]
percent_overall_passing = len(overall_passing)/ total_students * 100


# Create a dataframe to hold the above results
district_summary = {'Total Schools':  [total_schools],
        'Total Students': [total_students],
        'Total Budget': [total_budget],
        'Average Math Score': [avg_math_score],
        '% Passing Math': [percent_passing_math],
        'Average Reading Score': [avg_reading_score],
        '% Passing Reading': [percent_passing_reading],
        '% Overall Passing': [percent_overall_passing]
        }
district_summary_df = pd.DataFrame(district_summary)


# Formatting
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df
```

```python
# SCHOOL SUMMARY

# School type
school_type = school_data.set_index(['school_name'])["type"]

grouped_school_df = school_data_df.groupby(['school_name'])

# Students per school
stu_per_school = grouped_school_df["Student ID"].count()

# School budget
school_budget = grouped_school_df["budget"].unique()

# Per student budget
per_stu_budget = school_budget/stu_per_school

# Calculate the average math score
avg_school_mscore = grouped_school_df["math_score"].mean()

# # Calculate the average reading score
avg_school_rscore = grouped_school_df["reading_score"].mean()

# % Passing Math
grouped_passing_math = num_passing_math.groupby('school_name')
school_pass_math = (grouped_passing_math['Student ID'].count()/stu_per_school) * 100

# % Passing Reading

grouped_passing_reading = num_passing_reading.groupby('school_name')
school_pass_reading = (grouped_passing_reading['Student ID'].count()/stu_per_school) * 100

# % Overall passing
grouped_overall_passing = overall_passing.groupby('school_name')
school_overall_pass = (grouped_overall_passing['Student ID'].count()/stu_per_school) * 100

# Create dataframe
school_summary = {'Type': school_type,
        'Students Enrolled': stu_per_school,
        'School Budget': school_budget,
        'Per Student Budget': per_stu_budget,
        'Average Math Score': avg_school_mscore,
        '% Passing Math': school_pass_math,
        'Average Reading Score': avg_school_rscore,
        '% Passing Reading': school_pass_reading,
        '% Overall Passing': school_overall_pass
        }

school_summary_df = pd.DataFrame(school_summary)


# Remove list brackets from 'School Budget' column by grabbing the item with index 0 in each row
school_summary_df['School Budget'] = school_summary_df['School Budget'].str.get(0)
# Formatting
school_summary_df["School Budget"] = school_summary_df["School Budget"].map("${:,.2f}".format)

# Remove list brackets from 'Per Student Budget' column by grabbing the item with index 0 in each row
school_summary_df['Per Student Budget'] = school_summary_df['Per Student Budget'].str.get(0)
# Formatting
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)

school_summary_df
```

```python
# TOP PERFORMING SCHOOLS
top_schools = school_summary_df.sort_values("% Overall Passing", ascending=False)

top_schools.head(5)
```

```python
# BOTTOM PERFORMING SCHOOLS
worst_schools = school_summary_df.sort_values("% Overall Passing", ascending=True)

worst_schools.head(5)
```

```python
# MATH SCORES BY GRADE

# Freshmen average math score
all_freshmen = school_data_df[school_data_df["grade"]=="9th"]
grouped_fresh = all_freshmen.groupby("school_name")
fresh_avgm = grouped_fresh['math_score'].mean()

# Sophmore average math score
all_sophmores = school_data_df[school_data_df["grade"]=="10th"]
grouped_soph = all_sophmores.groupby("school_name")
soph_avgm = grouped_soph['math_score'].mean()

# Junior average math score
all_juniors = school_data_df[school_data_df["grade"]=="11th"]
grouped_jun = all_juniors.groupby("school_name")
jun_avgm = grouped_jun['math_score'].mean()

# Senior average math score
all_seniors = school_data_df[school_data_df["grade"]=="12th"]
grouped_sen = all_seniors.groupby("school_name")
sen_avgm = grouped_sen['math_score'].mean()

# Create dataframe
math_summary = {'9th Grade Average Math Score': fresh_avgm,
        '10th Grade Average Math Score': soph_avgm,
        '11th Grade Average Math Score': jun_avgm,
        '12th Grade Average Math Score': sen_avgm
        }

math_by_grade = pd.DataFrame(math_summary)

math_by_grade
```

```python
# READING SCORES BY GRADE

# Freshmen average reading score
fresh_avgr = grouped_fresh['reading_score'].mean()

# Sophmore average math score
soph_avgr = grouped_soph['reading_score'].mean()

# Junior average math score
jun_avgr = grouped_jun['reading_score'].mean()

# Senior average math score
sen_avgr = grouped_sen['reading_score'].mean()

# Create dataframe
reading_summary = {'9th Grade Average Reading Score': fresh_avgr,
        '10th Grade Average Reading Score': soph_avgr,
        '11th Grade Average Reading Score': jun_avgr,
        '12th Grade Average Reading Score': sen_avgr
        }

reading_by_grade = pd.DataFrame(reading_summary)

reading_by_grade
```

```python
# SCORES BY SCHOOL SPENDING

# Create basic dataframe
scores_by_spend = {'Per Student Budget': per_stu_budget,
        'Average Math Score': avg_school_mscore,
        '% Passing Math': school_pass_math,
        'Average Reading Score': avg_school_rscore,
        '% Passing Reading': school_pass_reading,
        '% Overall Passing': school_overall_pass
        }
scores_by_spend_df = pd.DataFrame(scores_by_spend)

# Remove list brackets from 'Per Student Budget' column by grabbing the item with index 0 in each row
scores_by_spend_df['Per Student Budget'] = scores_by_spend_df['Per Student Budget'].str.get(0)

# Create Bins
bins = [0, 585, 630, 645, 675]

# Create Bin names
spend_groups = ["<$584", "$585-629", "$630-644","$645-675"]

# Create Test Score Summary column
scores_by_spend_df["Spending Ranges (Per Student)"] = pd.cut(scores_by_spend_df["Per Student Budget"], bins, labels=spend_groups, include_lowest=True)

# Remove 'Per Student Budget' column (y/vertical axis = 0, x/horizontal axis = 1)
scores_by_spend_df = scores_by_spend_df.drop(["Per Student Budget"], axis=1)

# Group by the new spending ranges column
scores_by_spend_df = scores_by_spend_df.groupby('Spending Ranges (Per Student)')

# Find the average scores
scores_by_spend_df = scores_by_spend_df.mean()

# Formatting
scores_by_spend_df["% Overall Passing"] = scores_by_spend_df["% Overall Passing"].map("{:.2f}".format)
scores_by_spend_df["% Passing Reading"] = scores_by_spend_df["% Passing Reading"].map("{:.2f}".format)
scores_by_spend_df["% Passing Math"] = scores_by_spend_df["% Passing Math"].map("{:.2f}".format)
scores_by_spend_df["Average Reading Score"] = scores_by_spend_df["Average Reading Score"].map("{:.2f}".format)
scores_by_spend_df["Average Math Score"] = scores_by_spend_df["Average Math Score"].map("{:.2f}".format)

scores_by_spend_df
```

```python
# SCORES BY SCHOOL SIZE

# Create basic dataframe
scores_by_size = {'Students Enrolled': stu_per_school,
        'Average Math Score': avg_school_mscore,
        '% Passing Math': school_pass_math,
        'Average Reading Score': avg_school_rscore,
        '% Passing Reading': school_pass_reading,
        '% Overall Passing': school_overall_pass
        }
scores_by_size_df = pd.DataFrame(scores_by_size)

# Create Bins
bins = [0, 1000, 2000, 5000]

# Create Bin names
size_groups = ["Small (<1000)", "Medium (1000-1999)", "Large (2000-5000)"]

# Create Test Score Summary column
scores_by_size_df["School Size"] = pd.cut(scores_by_size_df["Students Enrolled"], bins, labels=size_groups, include_lowest=True)

# Remove 'Students Enrolled' column (y/vertical axis = 0, x/horizontal axis = 1)
scores_by_size_df = scores_by_size_df.drop(["Students Enrolled"], axis=1)

# Group by size
scores_by_size_df = scores_by_size_df.groupby('School Size')

# Find the average scores
scores_by_size_df = scores_by_size_df.mean()

# Formatting
scores_by_size_df["% Overall Passing"] = scores_by_size_df["% Overall Passing"].map("{:.2f}".format)
scores_by_size_df["% Passing Reading"] = scores_by_size_df["% Passing Reading"].map("{:.2f}".format)
scores_by_size_df["% Passing Math"] = scores_by_size_df["% Passing Math"].map("{:.2f}".format)
scores_by_size_df["Average Reading Score"] = scores_by_size_df["Average Reading Score"].map("{:.2f}".format)
scores_by_size_df["Average Math Score"] = scores_by_size_df["Average Math Score"].map("{:.2f}".format)

scores_by_size_df
```

```python
# SCORES BY SCHOOL TYPE

# Create basic dataframe
scores_by_type = {'School Type': school_type,
        'Average Math Score': avg_school_mscore,
        '% Passing Math': school_pass_math,
        'Average Reading Score': avg_school_rscore,
        '% Passing Reading': school_pass_reading,
        '% Overall Passing': school_overall_pass
        }
scores_by_type_df = pd.DataFrame(scores_by_type)

# Group by type
scores_by_type_df = scores_by_type_df.groupby('School Type')

# Find the average scores
scores_by_type_df = scores_by_type_df.mean()

# Formatting
scores_by_type_df["% Overall Passing"] = scores_by_type_df["% Overall Passing"].map("{:.2f}".format)
scores_by_type_df["% Passing Reading"] = scores_by_type_df["% Passing Reading"].map("{:.2f}".format)
scores_by_type_df["% Passing Math"] = scores_by_type_df["% Passing Math"].map("{:.2f}".format)
scores_by_type_df["Average Reading Score"] = scores_by_type_df["Average Reading Score"].map("{:.2f}".format)
scores_by_type_df["Average Math Score"] = scores_by_type_df["Average Math Score"].map("{:.2f}".format)

scores_by_type_df
```
